```python
import numpy as np
import matplotlib.pyplot as plt

#defineiteration
iteration = 100

#definesimulationspace
xmin = 0
xmax = 120
ymax = 120
ymin = 0
xrange = xmax - xmin
yrange = ymax - ymin
pos = []
posisi = []

#defineinitialposition
n_particles = 10 #definetotalparticle
for i in range(n_particles):
  x_pos = np.random.randint(low=xmin, high=xmax)
  y_pos = np.random.randint(low=ymin, high=ymax)
  pos.append([x_pos, y_pos])
posisi.append(pos)

  
for i in range(iteration):
    change_position = []
    #printiterasi
    print('Iteration',i+1)
    for j in range(n_particles):
        pos_x = posisi[i][j][0]
        pos_y = posisi[i][j][1]
        #updateposition
        rand = np.random.rand()
        #right
        if(rand<=0.25):
            pos_x = pos_x + 1
        #down
        elif(rand<=0.50):
            pos_y = pos_y - 1
        #left
        elif(rand<=0.75):
            pos_x = pos_x - 1
        #up
        else:
            pos_y = pos_y + 1
        #performpbccorrection
        if (pos_x > xmax): 
            pos_x = pos_x - xrange
        if (pos_x < 0): 
            pos_x = pos_x + xrange
        if (pos_y > ymax):
            pos_y = pos_y - yrange
        if (pos_y < 0): 
            pos_y = pos_y + yrange
        #printcekperpindahanposisi
        print('Particle',j+1,pos_x,pos_y)
        change_position.append([pos_x,pos_y])
    posisi.append(change_position)
   
fig, ax = plt.subplots(figsize=(10,10))
for i in range(iteration):
  for j in range(n_particles):
    ax.scatter(posisi[i][j][0], posisi[i][j][1], s=50, alpha=0.75)
plt.show()
```
